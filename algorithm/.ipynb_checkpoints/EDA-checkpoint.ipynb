```python
import pandas as pd
import numpy as np
# 绘图
import matplotlib
from matplotlib import pyplot as plt 
matplotlib.rcParams['font.sans-serif'] = ['SimHei'] # 用黑体显示中文
```

```python
train = pd.read_csv("../data/used_car_train_20200313.csv")
test = pd.read_csv("../data/used_car_testA_20200313.csv")
# print(train.head(),test.head())
print(train.shape)
print(train.info())
train_data_size = train.shape[0]
```

```python
# 成交价格分布
price_list = train['price']
price_max = price_list.max()
price_min = price_list.min()
# print(price_min,price_max)
interval = int((price_max) / 20 +1)

price_y = []
price_x = []

for index in range(20) :
    price_y.append(0)
    price_x.append(index+1)
# print(price_y,price_x)
for price in price_list:
    price_y[ int(price/interval) ]+=1
print(price_y)
    
plt.bar(x= price_x,height = price_y ,label ="价格分布",color = 'steelblue',alpha=0.8)
for x , y in zip(price_x,price_y):
    plt.text(x,y+1,str(y),ha='center',va='bottom',fontsize=10, rotation=0)
plt.title("二手车成交价格分布图(单位5000)")
plt.xlabel("成交价格%d"%(interval))
plt.ylabel("成交次数")
plt.legend()
plt.savefig("../fig/price5000.jpg")
plt.show()
```

```python
# 数据缺失情况
na_count =  train.isna().sum()

# print(na_count)
reserve_index = []
for index in na_count.index:
    if na_count[index] != 0:
        reserve_index.append(index)
na_count=na_count[reserve_index]
print(na_count)
na_count_x =  list(na_count.index)
na_count = na_count /train_data_size*100
na_count_y = list(na_count)



plt.bar(x= na_count_x,height = na_count_y ,label ="缺失情况",color = 'steelblue',alpha=1)
for x , y in zip(na_count_x,na_count_y):
    plt.text(x,y+1,str(y),ha='center',va='bottom',fontsize=10, rotation=0)
plt.title("数据缺失情况")
plt.xlabel("列名")
plt.ylabel("缺失比百分比")
plt.legend()
plt.savefig("../fig/na.jpg")
plt.show()

# print(na_count_x,na_count_y)
```
